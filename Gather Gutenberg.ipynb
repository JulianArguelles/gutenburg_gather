#### Where you want to save text to

```python
save_path = 'C:\Users\chari\Desktop\Word2Vec\gutenberg_english_fantasy_corpus'
```

#### Gather all plaintext of english books with subject fantasy or fiction.

```python
for book in range(1, 20000):
    #False if id does frozenset is empty
    exists = True
    
    #True if subject contains fantasy or fiction
    fiction_or_fantasy = False
    
    #Frozensets of name, language, and subject of the book id
    name = get_metadata('title', book)
    language = get_metadata('language', book)
    subject = get_metadata('subject', book)
    
    #Returns language as a string
    try:
        language = iter(language).next().encode('ascii','ignore')
        pass
    except StopIteration:
        pass
    
    #If language is not English, go back to the beginning
    if "en" not in language:
        continue

    #Makes fiction_or_fantasy = True if subject contains fiction or fantasy
    for sub in subject:
        if "fantasy" in sub.lower():
            fiction_or_fantasy = True
          
    for sub in subject:
        if "fiction" in sub.lower():
            fiction_or_fantasy = True

    #If book is not fiction or fantasy, go back to the beginning
    if not fiction_or_fantasy:
        continue
    
    #Turns name into a string, if DNE, go back to the beginning
    try:
        name = iter(name).next().encode('ascii','ignore')
    except StopIteration:
        exists = False
    
    if exists:
        #Name of the file
        name_of_file = name
        
        #File path of a book
        completeName = os.path.join(save_path, name_of_file + ".txt")
        
        #Load text with headers stripped
        try:
            text = strip_headers(load_etext(book)).strip()
        except:
            continue
        
        #Ignore any ascii (\u) characters, no need to replace in this case
        text = text.encode('ascii','ignore')
        
        #Replace any instances of \r\n with a space
        text = text.replace("\\r\\n", " ")
        
        if not text:
            continue
        
        #Checks to see if book already exists
        if(os.path.exists(completeName) == False):

            try:
                file1 = open(completeName, "w")
                pass
            except:
                continue

            toFile = text

            file1.write(toFile)

            file1.close()
        
    else:
        continue
```

#### Get rid of any bad files

```python
deleter = os.listdir(save_path)

for item in deleter:
    if item.endswith(".txt"):
        pass
    else:
        try:
            os.remove(save_path + "\\" +  item)
        except:
            pass
```
